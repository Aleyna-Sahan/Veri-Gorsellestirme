```python
import matplotlib.pyplot as plt
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

data_path = '/content/drive/MyDrive/excel/piramit.xlsx'
df = pd.read_excel(data_path)

print(df.columns)

plt.figure(figsize=(10, 8))

from pandas.plotting import parallel_coordinates

df_parallel = df[['İL','Yıl_2002', 'Yıl_2003', 'Yıl_2004', 'Yıl_2005', 'Yıl_2006']]

parallel_coordinates(df_parallel,'İL', colormap='turbo', linewidth=2)

plt.xlabel('Yıllar', fontsize=12, fontweight='bold')
plt.ylabel('Değerler', fontsize=12, fontweight='bold')
plt.title('Paralel Koordinatlar Grafiği - Yıllar Bazında İl ve Yıl Düzeyi Kimyevi Gübre Tüketim Verileri', fontsize=14, fontweight='bold')

plt.xticks(rotation=45, fontsize=10)

plt.grid(True, linestyle='-', alpha=0.5)

plt.gca().set_facecolor('#d6f9de')

plt.show()
```
