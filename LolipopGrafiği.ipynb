```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt
file_path = '/content/drive/My Drive/excel/lolipop.xlsx'
data = pd.read_excel(file_path)
plt.figure(figsize=(12, 6), facecolor='lightblue')
categories = data['İL']
values = data['Toplam AZOT(N)']
plt.stem(categories, values, linefmt='--', markerfmt='o', basefmt='-', use_line_collection=True)
plt.xlabel('İL')
plt.ylabel('Toplam AZOT(N)')
plt.title('İller ve Azot Miktarı')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
