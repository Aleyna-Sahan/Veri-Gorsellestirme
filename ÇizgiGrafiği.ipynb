```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
file_path = '/content/drive/My Drive/toplam.xlsx'
data = pd.read_excel(file_path)
print(data.head())
plt.figure(figsize=(14, 8))
years = data['İL']
production = data['Toplam TOPLAM BİTKİ BESİN MADDESİ ']
plt.plot(years, production, marker='o', linestyle='-', color='green', markerfacecolor='brown', markeredgewidth=2, markeredgecolor='darkgreen')
plt.xlabel('İL', fontsize=14, color='darkgreen')
plt.ylabel('İLLERE GÖRE TOPLAM BİTKİ BESİN MADDESİ', fontsize=14, color='darkgreen')
plt.title('İllere Göre Toplam Bitki Besin Maddesi', fontsize=16, color='darkgreen')
plt.xticks(rotation=90, fontsize=8)
plt.grid(True, linestyle='-', linewidth=0.7, color='gray', alpha=0.5)
plt.gca().set_facecolor('floralwhite')
plt.tight_layout()
plt.show()
```
