```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
file_path = '/content/drive/My Drive/excel/marmara.xlsx'
df = pd.read_excel(file_path)
kategori = df['Marmara']
deger = df['Gübre Tüketimi']
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(aspect="equal"))
wedges, texts, autotexts = ax.pie(deger, labels=kategori, autopct='%1.1f%%', startangle=140, colors=plt.cm.viridis(np.linspace(0, 1, len(kategori))))
for text, wedge in zip(texts, wedges):

    center = wedge.center


    x, y = center[0], center[1]


    if x > 0:  # Sağdaki dilimler için
        rotation = np.degrees(np.arctan2(y, x))
        text.set_rotation(rotation)
        text.set_horizontalalignment('right')  # Sağa hizalama
    else:  # Soldaki dilimler için
        rotation = np.degrees(np.arctan2(y, x)) + 180
        text.set_rotation(rotation)
        text.set_horizontalalignment('left')  # Sola hizalama

    # Ters çevirme
    text.set_rotation(0)  # 180 derece (ters) döndürme


plt.title('2000 Yılındaki İl Verileri Pasta Grafiği')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
