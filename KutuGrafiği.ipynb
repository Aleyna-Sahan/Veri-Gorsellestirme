```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
file_path = '/content/drive/MyDrive/excel/kutu.xlsx'
df = pd.read_excel(file_path)
numeric_df = df.select_dtypes(include=['number'])
colors = plt.cm.viridis(np.linspace(0, 1, len(numeric_df.columns)))
plt.figure(figsize=(12, 8))
plt.title('Veri Kutu Grafiği', fontsize=16)
plt.xlabel('Sütunlar', fontsize=14)
plt.ylabel('Değerler', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
box = plt.boxplot(numeric_df.values, patch_artist=True, labels=numeric_df.columns)
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
plt.show()
```
