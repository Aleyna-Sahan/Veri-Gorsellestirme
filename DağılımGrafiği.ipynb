```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
file_path = '/content/drive/My Drive/excel/Dağılım.xlsx'
data = pd.read_excel(file_path)
print(data.head())
plt.figure(figsize=(10, 6))
azot = data['İL']
fosfor = data['Grand Total']
plt.scatter(azot, fosfor, color='green')
plt.xlabel('İL')
plt.xticks(rotation=89, fontsize=8)
plt.ylabel('Grand Total')
plt.title('Toplam Azot ve Fosfor İlişkisi')
plt.grid(True)
plt.tight_layout()
plt.show()
```
