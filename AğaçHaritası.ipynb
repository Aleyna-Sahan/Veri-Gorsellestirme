```python
from google.colab import drive
drive.mount('/content/drive')
!pip install squarify
import pandas as pd
import matplotlib.pyplot as plt
import squarify
file_path = '/content/drive/My Drive/marmara.xlsx'
df = pd.read_excel(file_path)
df.head()
categories = df['Marmara'].tolist()
values = df['Gübre Tüketimi'].tolist()
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(12, 8)
cmap = plt.cm.viridis
norm = plt.Normalize(vmin=min(values), vmax=max(values))
colors = [cmap(norm(value)) for value in values]
squarify.plot(sizes=values, label=categories, color=colors, alpha=0.8, ax=ax)
plt.title('Ağaç Haritası', size=20)
plt.axis('off')
plt.show()
```
