```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
file_path = '/content/drive/My Drive/excel/Bar.xlsx'
data = pd.read_excel(file_path)
print(data.head())
data_sorted = data.sort_values(by='Toplam AZOT(N)', ascending=True)
categories = data_sorted['İL']
values = data_sorted['Toplam AZOT(N)']
colors = plt.cm.Purples(np.linspace(0, 1, len(values)))
plt.style.use('seaborn-poster')
plt.figure(figsize=(20, 6))
arka_plan_resmi = plt.imread('/content/drive/My Drive/excel/Bar.jpg')
plt.imshow(arka_plan_resmi, extent=[-1, len(categories), 0, max(values)*1.1], aspect='auto', alpha=0.5)
plt.bar(categories, values, color=colors, edgecolor='purple')
plt.xlabel('İL', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Azot', fontsize=14, fontweight='bold', color='darkblue')
plt.title('İl ve Azot Bar Grafiği', fontsize=18, fontweight='bold', color='darkblue')
plt.xticks(rotation=90, fontsize=12, color='darkblue')
plt.yticks(fontsize=12, color='darkblue')
plt.gca().set_facecolor('none')
plt.tight_layout()
plt.show()
```
