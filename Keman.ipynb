```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
data_path = '/content/drive/MyDrive/excel/piramit.xlsx'
df = pd.read_excel(data_path)
df_melted = pd.melt(df, id_vars=['İL'], value_vars=['Yıl_2000', 'Yıl_2005', 'Yıl_2010', 'Yıl_2015'],
                    var_name='Yıl', value_name='Yıllar')
plt.figure(figsize=(15, 8))
sns.violinplot(x='Yıl', y='Yıllar', data=df_melted, inner='quartile', linewidth=1.3, palette='mako')
plt.xlabel('Yıllar', fontsize=12, fontweight='bold')
plt.ylabel('Ödünç Verilen Materyal Sayısı', fontsize=12, fontweight='bold')
plt.title('Ödünç Verilen Materyal Keman Grafiği', fontsize=14, fontweight='bold')
plt.xticks(rotation=90, fontsize=8)
plt.yticks(fontsize=8)
plt.grid(True, axis='y', linestyle='-', alpha=0.5)
plt.gca().set_facecolor('#cbf4d8')
plt.show()
```
