```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
file_path = '/content/drive/My Drive/excel/sütun.xlsx'
data = pd.read_excel(file_path)
print(data.head())
arka_plan_resmi = plt.imread('/content/drive/My Drive/excel/Yeşil.jpg')
fig, ax = plt.subplots(figsize=(14, 8))
ax.imshow(arka_plan_resmi, extent=[-1, len(data['İL']), 0, max(data['Toplam POTAS (K2O)'])*1.1], aspect='auto', alpha=0.5)
ax.bar(data['İL'], data['Toplam POTAS (K2O)'], color='green', edgecolor='black')
ax.set_xlabel('İL', fontsize=14, color='darkgreen')
ax.set_ylabel('Toplam POTAS (K2O)', fontsize=14, color='darkgreen')
ax.set_title('İllere Göre Potasyum (P2O5) Miktarı', fontsize=16, color='darkgreen')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
